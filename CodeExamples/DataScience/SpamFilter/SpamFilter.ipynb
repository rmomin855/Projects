```python
import os
import re
import numpy as np
import string
```

<p>Our first step is to take the downloaded data parse out all the subject headers.</p>
<ol>
    <li>First we loop through each raw email file.</li>
    <li>Then we check if the raw email data contains "Subject:"</li>
    <li>We then convert the data to 'utf-8' string format, removing any non-complient headers.</li>
    <li>We also remove all punctuation and set each string to lower-case for easy comparisons.</li>
</ol>

```python
def parse_folder_to_list(folder):
    list_of_subjects = []
    for filename in os.listdir(folder):
        with open(os.path.join(os.path.dirname(folder), filename), 'rb', 0) as search:
            for line in search:
                line = line.rstrip()
                if re.search(br'Subject:', line):
                    try:
                        list_of_subjects.append(line[8:].decode().translate(str.maketrans('', '', string.punctuation)).lower())
                    except Exception:
                        #print("File: ", filename, " --- Failed to decode: ", line)
                pass
        pass
    return list_of_subjects
```

```python
spam = parse_folder_to_list('./spam/')

easy_ham = parse_folder_to_list('./easy_ham/')
hard_ham = parse_folder_to_list('./hard_ham/')
ham = easy_ham + hard_ham
```

Next, we split our data into 75% Training and 25% Testing. This will be used to train and test our spam filter.

```python
from sklearn.model_selection import train_test_split
spam_train, spam_test = train_test_split(spam, test_size=0.25)
ham_train, ham_test = train_test_split(ham, test_size=0.25)
print("Spam/Ham Training data size:", len(spam_train), "/", len(ham_train))
print("Spam/Ham Training data size:", len(spam_test), "/", len(ham_test))
```

Now we take the Training data and create a list of unique words to be used as a dictionary.

```python
spam_dic = {}
ham_dic = {}

for subjectHeader in spam_train:
    for x in subjectHeader.split():
        try:
            spam_dic[x] += 1
        except Exception:
            spam_dic[x] = 1
        pass
    pass

for subjectHeader in ham_train:
    for x in subjectHeader.split():
        try:
            ham_dic[x] += 1
        except Exception:
            ham_dic[x] = 1
        pass
    pass

total_dic = dict(spam_dic);
total_dic.update(ham_dic)
print("Number of words in our dictionary: ", len(spam_dic), " / ", len(ham_dic))
print("Total unique words: ", len(total_dic))
```

Now we may begin working on our Spam Filter!

```python
#lets define our constants:
numSpam = len(spam_dic)
numHam = len(ham_dic)
numWords = len(total_dic)
prob_spam = numSpam / numWords
prob_ham = numHam / numWords
alpha = 1
beta = 2
```

```python
#Createa a function that caluclates the prob of being spam
def prob_of_word_given_spam(word):
    try:
        value = spam_dic[word]
    except Exception:
        value = 0
    return (value + alpha)/ (numSpam + beta)
def prob_of_word_given_ham(word):
    try:
        value = ham_dic[word]
    except Exception:
        value = 0
    return (value + alpha)/ (numHam + beta)

def prob_of_spam_given_word(word):
    numerator = prob_of_word_given_spam(word)
    denominator = numerator + prob_of_word_given_ham(word)
    return numerator / denominator
```

```python
word_matrix = []
for word in total_dic:
    word_matrix.append(prob_of_spam_given_word(word))
```
